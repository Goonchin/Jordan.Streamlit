## As you can see we downloaded 100 pictures of each shoe model that was listed from Bing search engine.







```python
brands = DataBlock(blocks=(ImageBlock, CategoryBlock),
                    get_items=get_image_files,
                    splitter=RandomSplitter(valid_pct=0.2, seed=42),
                    get_y=parent_label,
                    item_tfms=RandomResizedCrop(224, min_scale=0.5), 
                    batch_tfms=aug_transforms())
```

```python
path = Path("/content/drive/MyDrive/Sneakers /Brands")
```

```python
dls = brands.dataloaders(path)
```

```python
dls.valid.show_batch(max_n=8, nrows=2)
```

## Preform Transfer learning

```python
learn = cnn_learner(dls, resnet18, metrics=error_rate)
learn.fine_tune(10)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```

```python
interp.plot_top_losses(15, nrows=3)
```
